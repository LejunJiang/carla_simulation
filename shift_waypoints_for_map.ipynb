```python
import numpy as np
import matplotlib.pyplot as plt

npzfile = np.load("map.npz")
x_map = npzfile['arr_0']
y_map = npzfile['arr_1']
x_center = np.mean(x_map)
y_center = np.mean(y_map)
waypoints_map = npzfile['arr_2']
print(x_map.shape)
print(waypoints_map.shape)
plt.figure()
plt.plot(x_map, y_map, '.')
plt.plot(x_center, y_center, '.')
plt.show()
```

```python
from copy import deepcopy
x_map_new = deepcopy(x_map)
y_map_new = deepcopy(y_map)
waypoints_map_new = deepcopy(waypoints_map)
for i in range(0, x_map.shape[0]):
    direction_vector = np.array([x_center, y_center]) - waypoints_map[i, :]
    waypoints_map_new[i, :] = waypoints_map[i, :] + 0.8 * direction_vector / np.linalg.norm(direction_vector)
    x_map_new[i] = waypoints_map_new[i, 0]
    y_map_new[i] = waypoints_map_new[i, 1]
plt.figure()
plt.plot(x_map, y_map, '.')
plt.plot(x_map_new, y_map_new, '.')
np.savez("map_shifted_waypoints.npz", x_map_new, y_map_new, waypoints_map_new)
plt.show()
```
